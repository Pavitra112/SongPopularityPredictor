```python
def find_outliers(doc,col):
    q1 = doc[col].quantile(0.25)
    q3 = doc[col].quantile(0.75)
    IQR = q3 - q1
    low_limit = q1 - IQR*1.5
    high_limit = q3 + IQR*1.5
    doc = doc.loc[(doc[col]<low_limit)|(doc[col]>high_limit)]
    return doc
find_outliers(df,'bpm')
```

```python
def remove_outliers(doc,col):
    q1 = doc[col].quantile(0.25)
    q3 = doc[col].quantile(0.75)
    IQR = q3 - q1
    low_limit = q1 - IQR*1.5
    high_limit = q3 + IQR*1.5
    doc = doc.loc[(doc[col]>=low_limit)&(doc[col]<=high_limit)]
    return doc
df1 = remove_outliers(df,'bpm')
```

```python
df1
```

```python
# value of x should always dataframe.
x = df1.drop(['pop','title','artist','top genre','year'],axis=1)
x
```

```python
# value of y dataframe or array.
y = df1['pop']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=29)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)# making equation.
print(lr.coef_) # value of coeficiente (b) in equation.
import numpy as np
print(np.round(lr.coef_))
print(lr.intercept_) # value of intercept (a)
```

```python
x.columns
```

```python
y_pred = lr.predict(x_test)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_pred,y_test))
```
